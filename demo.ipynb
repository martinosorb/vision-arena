```python
from arena import Arena
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# define a function that associates every point in [0, 2) to a colour
color_function = lambda angle: plt.cm.jet(angle/2/np.pi)

a = Arena(color_function,
          visual_breadth=np.pi/8,  # aperture of the visual field (rad)
          visual_resolution=100)    # number of pixels recorded by the agent

# plot with default location and orientation
a.plot_combined()

# set location and orientation
# the field of view is automatically updated
a.pos = (0.5, 0.5)
a.phi = np.pi*3/4

a.plot_combined()
```

```python
# you can access the angles corresponding to what is seen
print(a.vf_angles)

# or the colors
print(a.vf_colors)
```
